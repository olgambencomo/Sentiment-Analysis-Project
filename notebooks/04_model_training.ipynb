# Model Training & Selection

Train and compare machine learning models for sentiment classification.

Models:
- Logistic Regression
- Support Vector Machine (SVM)

Techniques:
- TF-IDF vectorization
- Stratified K-Fold Cross Validation
- Grid Search hyperparameter tuning

Evaluation Metric:
- F1-score (macro) to handle class imbalance

```python
# Import Libraries

import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report
```

```python
#Load the data

train = pd.read_csv('/Users/olgabencomo/Desktop/Proyectos Portafolio/Sentiment Analysis Project/data/processed/train.csv')
val = pd.read_csv('/Users/olgabencomo/Desktop/Proyectos Portafolio/Sentiment Analysis Project/data/processed/val.csv')

X_train = train['text']
y_train = train['label']

X_val = val['text']
y_val = val['label']
```

```python
# Pipeline

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('model', LogisticRegression())  # placeholder
])
```

```python
# Stratified K-Fold

cv = StratifiedKFold(
    n_splits=3,
    shuffle=True,
    random_state=42
)
```

```python
# Hyperparameter Grid


param_grid = [

    #  Logistic Regression
    {
        'model': [LogisticRegression(max_iter=300, class_weight='balanced')],
        'tfidf__max_features': [3000, 5000],
        'tfidf__ngram_range': [(1,1), (1,2)],
        'model__C': [0.1, 1, 10]
    },

    #  SVM
    {
        'model': [LinearSVC(class_weight='balanced')],
        'tfidf__max_features': [3000, 5000],
        'tfidf__ngram_range': [(1,1), (1,2)],
        'model__C': [0.1, 1, 10]
    }
]
```

```python
# Grid Search

grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=cv,
    scoring='f1_macro',
    n_jobs=-1,
    verbose=2
)
```

```python
# Training

grid.fit(X_train, y_train)
```

```python
# Best Parameters and Score

print("Best Parameters:")
print(grid.best_params_)

print("\nBest CV F1 Score:")
print(grid.best_score_)
```

```python
best_model = grid.best_estimator_
```

```python

# Validation Performance

y_val_pred = best_model.predict(X_val)

print(classification_report(y_val, y_val_pred))
```

```python
# Best Results


results = pd.DataFrame(grid.cv_results_)

results = results.sort_values(by='mean_test_score', ascending=False)

results[['params', 'mean_test_score']].head(10)
```

## Model Training Conclusion

In this stage, multiple machine learning models were trained and evaluated for sentiment classification, including Logistic Regression and Support Vector Machine (SVM).

To create the target variable, customer ratings were used as a proxy for sentiment:
- Ratings ≤ 2 → Negative  
- Rating = 3 → Neutral  
- Ratings ≥ 4 → Positive  

The best-performing model was a Linear SVM with TF-IDF features, optimized using GridSearchCV with Stratified K-Fold cross-validation and F1-score (macro) as the evaluation metric.

While the overall performance was strong, especially for the Positive class, the model showed lower performance for the Neutral class, which is inherently more ambiguous.

### Key Limitation

Using the rating to label sentiment introduces potential noise into the dataset.

This is because:
- A high rating does not always imply strongly positive language
- A neutral review (text) can still have a high or low rating depending on user bias
- The textual sentiment and numerical rating are not always perfectly aligned

### Key Takeaway

Although this labeling approach allows for supervised learning at scale, it may impact model performance due to label noise.

Future improvements could include:
- Manual labeling or curated datasets
- Combining rating-based labels with NLP-based sentiment signals
- Using transformer models to better capture nuanced sentiment

Overall, the SVM model provided a strong baseline with good generalization and efficiency.
